```python
#converting into array
password_tuple=np.array(df)
```

```python
#shuffling array
import random
random.shuffle(password_tuple)
```

```python
#independent and dependent splitting
X=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

```python
X
```

```python
y
```

```python
#applying tf-idf
```

```python
def word_divide_char(inputs):
  character=[]
  for i in inputs:
    character.append(i)
  return character  
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X_vec = tfidf.fit_transform(X)
```

```python
X_vec.shape
```

```python
tfidf.get_feature_names()
```

```python
first_document_vector = X_vec[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df2 = pd.DataFrame(first_document_vector.T.todense(),index=tfidf.get_feature_names(),columns=['TF=IDF'])
```

```python
df2
```

```python
df2.sort_values(by=['TF=IDF'],ascending=False)
```

```python
#Applying logistic regression algorithm
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_vec,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
log = LogisticRegression(random_state=42,multi_class='multinomial')  
```

```python
log.fit(X_train,y_train)
```

```python
test = np.array(['%@123abcdef'])
```

```python
pred = tfidf.transform(test)
log.predict(pred)
```

```python
y_pred = log.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
